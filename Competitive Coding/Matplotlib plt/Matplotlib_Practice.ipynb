# Ans). Box Plot

The max value is 5 because the third quartile is 4.5 and the interquartile range is (4.5-1.5)=>3. 

So, 1.5*3 is 4.5 and third quartile(4.5)+4.5=>9. So 10 is larger than the limit 9, thus it becomes an outlier.

```python
list_1 = [0, 1, 2, 3, 4, 5, 10]
plt.boxplot(list_1)
plt.show()
```
